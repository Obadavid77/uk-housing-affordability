## Summary of Key Findings

### Finding 1 — No Region in England is Affordable
In 2024 not a single region in England meets the internationally 
recognised affordability benchmark of 3.0x median annual salary. 
The England average is 7.7x. London is the worst at 12.4x. Even 
the most affordable region — the North East — requires 4.7 years 
of full salary to buy a home.

### Finding 2 — Affordability Has Worsened Over 10 Years
Between 2015 and 2024 affordability deteriorated in 6 out of 9 
regions. The North West worsened by 13.9%, East Midlands by 10.1%, 
and West Midlands by 9.0%. Only London, South East and East of 
England saw modest improvement as Southern house price growth 
slowed while wages rose.

### Finding 3 — The Crisis is Spreading North
The regions deteriorating fastest are not the most expensive ones. 
They are the mid-priced Northern and Midlands regions where workers 
believed they could still afford to buy. The housing crisis is no 
longer a London problem — it is an England-wide structural failure.

### Finding 4 — The South East is the Next Pressure Point
The Random Forest model (R²=0.887) predicts that if current trends 
continue the South East will reach 11.4x by 2028 — a 21% 
deterioration from 2024. London is projected to plateau at 13.2x. 
No region is forecast to reach the 3.0x benchmark at any point 
before 2028.

### Finding 5 — Affordability is Non-Linear and Regionally Specific
Linear Regression failed completely (R²=0.015) confirming that 
housing affordability dynamics cannot be captured by simple trend 
lines. Regional identity, momentum, and prior year performance are 
the strongest predictors — meaning each region follows its own 
trajectory and requires its own policy response.

## Business Recommendations

### Recommendation 1 — Prioritise Housing Supply in the North and Midlands
Government housing policy has historically focused on London and the 
South East. The data shows this is no longer where the crisis is 
growing fastest. The North West, East Midlands and West Midlands 
are deteriorating at a rate that will price out working families 
within a generation. New housing supply should be prioritised where 
affordability is worsening fastest — not where it is already worst.

### Recommendation 2 — Use Affordability Ratios in Mortgage Lending
Banks and mortgage lenders currently use loan-to-income multiples 
as their primary affordability measure. This analysis shows that 
the regional affordability ratio — house price divided by median 
regional earnings — is a more meaningful indicator of structural 
risk. Lenders should incorporate regional affordability ratios 
into their risk assessment frameworks, particularly for first-time 
buyers in the North West and Midlands where ratios are rising fastest.

### Recommendation 3 — Deploy Predictive Monitoring for Policy Planning
This project demonstrates that housing affordability can be modelled 
and forecast with reasonable accuracy using publicly available data. 
The government's Department for Levelling Up, Housing and Communities 
(DLUHC) and the Office for Budget Responsibility (OBR) should 
maintain a live affordability ratio dashboard — updated annually 
with ASHE and HPI data — to track whether policy interventions 
are having measurable impact. The model built in this project 
provides a starting framework for that system.

### Recommendation 4 — Target First-Time Buyer Support Regionally
Current first-time buyer schemes such as Help to Buy and the 
Lifetime ISA apply nationally with the same thresholds regardless 
of region. The data shows affordability challenges are highly 
regional. Support thresholds, property price caps, and eligibility 
criteria should be adjusted by region to reflect where affordability 
pressure is actually greatest and growing fastest.

## Limitations and Further Research

### Limitations of This Analysis
- **England regions only** — this analysis covers 9 English regions. 
  Scotland, Wales and Northern Ireland follow different housing market 
  dynamics and are not included.
- **Annual data only** — ASHE earnings data is published annually. 
  Monthly house price movements cannot be fully captured.
- **Median earnings used** — median full-time earnings represent the 
  typical worker but do not capture inequality within regions. A high 
  earner in the North East and a low earner in London both sit behind 
  the same regional median.
- **No mortgage rate data** — interest rates significantly affect 
  affordability in practice. The ratio used here assumes a cash 
  purchase which no first-time buyer makes. A mortgage-adjusted 
  affordability ratio would be more precise.
- **Projections assume current trends** — the 2025-2028 forecasts 
  assume no major policy changes, economic shocks or structural 
  market shifts. They should be treated as directional indicators 
  not definitive forecasts.

### Suggested Further Research
- Incorporate mortgage rate data to build a payment-to-income 
  affordability measure alongside the price-to-earnings ratio
- Extend to local authority level to identify specific areas within 
  regions that are most at risk
- Add housing supply data (new build completions by region) to test 
  whether supply increases reduce the affordability ratio
- Model the impact of specific policy interventions — for example 
  what level of new supply would be required to bring each region 
  back to a ratio of 5.0x within 10 years
